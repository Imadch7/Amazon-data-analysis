```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("amazon.csv")
```

```python
list(df.columns)
```

```python
df = df.drop_duplicates()
```

```python
dropped_columns = ['category','about_product','user_id','user_name','review_id','review_title','review_content','img_link','product_link']
df = df.drop(columns=dropped_columns)
```

```python
df.head(8)
```

```python
df['product_name'] = df['product_name'].apply(lambda x: x.split(' ')[0])
```

```python
df.head(8)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
df
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors = 'coerce')
df
```

```python
df['rating_count'].fillna(df['rating_count'].mean(), inplace=True)
```

```python
df
```

```python
list(df.dtypes)
```

```python
df['actual_price'] = df['actual_price'].str.replace(',', '')
df
```

```python
df.to_csv('cleaned_amazon_data.csv', index=False)
```
